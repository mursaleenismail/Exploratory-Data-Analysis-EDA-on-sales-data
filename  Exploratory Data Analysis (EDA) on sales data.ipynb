**Distribution**

Distribution refers to how the data values are spread across a given range. It helps determine the shape of the data, such as whether it is skewed, normally distributed, or concentrated in specific intervals.

**Central Tendency**

Central tendency describes the midpoint or typical value in the dataset, commonly measured using the mean, median, or mode. It provides insight into the overall average behavior of the data.

**Variability**

Variability indicates how much the data values fluctuate around the central point. Measures such as range, variance, and standard deviation help assess the degree of dispersion in the dataset.

**Outliers**

Outliers are data points that significantly deviate from the majority of values. These unusual observations may result from data entry errors, rare events, or natural variability, and they can impact the accuracy of the analysis if not properly addressed.

#### Skewness and Kurtosis
**Skewness**

Skewness measures the asymmetry of the data distribution:

 - A skewness of 0 indicates a perfectly symmetrical distribution.

 - A positive skewness indicates a right-skewed distribution, where the tail extends toward higher values.

 - A negative skewness indicates a left-skewed distribution, where the tail extends toward lower values.

**Kurtosis**

Kurtosis describes the â€œtailednessâ€ of the distribution, indicating the presence of outliers:

 - A kurtosis of 3 suggests a normal distribution with a standard level of tails.

 - A kurtosis greater than 3 indicates a distribution with heavier tails and more extreme values.

 - A kurtosis less than 3 suggests a flatter distribution with fewer extreme values.

**Detect Outliers Using Z-Score Method**

```python
import numpy as np
from scipy import stats

numeric_cols = ['Sales', 'Profit', 'Discount', 'Quantity']

z_scores = np.abs(stats.zscore(dt[numeric_cols]))

threshold = 3

outliers = (z_scores > threshold)

outliers_sum = np.sum(outliers, axis=0)
print("Number of Outliers per Column (Z-Score Method):")
print(outliers_sum)
```

**IQR Method**

```python
Q1 = dt[numeric_cols].quantile(0.25)
Q3 = dt[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = ((dt[numeric_cols] < lower_bound) | (dt[numeric_cols] > upper_bound))

print("Number of Outliers per Column (IQR Method):")
print(outliers_iqr.sum())
```

**Step 6 Handling Outliers**

#### Identifying Outliers
#### Interquartile Range (IQR) Method

```python
for col in numeric_cols:
    q1 = dt[col].quantile(0.25)
    q3 = dt[col].quantile(0.75)
    iqr = q3 - q1
    lower=q1 - iqr*1.5
    upper=q3 + iqr*1.5

    outlier = ((dt[col] < lower) | (dt[col] > upper)).sum()
    
    print( f"OUTLIERS IN {col} : {outlier}" )
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(dt[numeric_cols]))
outliers = (z_scores > 3)
outlier_counts = np.sum(outliers, axis=0)
print("Outliers (Z-Score Method):")
print(dict(zip(numeric_cols, outlier_counts)))
```

**Since the Sales and Profit variables are not normally distributed and exhibit right-skewness, the Interquartile Range (IQR) method is more appropriate for detecting outliers. The Z-score method is typically applied when the data follows a normal distribution; therefore, IQR provides a more reliable approach for this dataset.**

**Deciding How to Handle Outliers**

We retained the outliers because they represent high-value transactions rather than data errors. These values reflect genuine business performance and help identify top-selling regions, products, or customers. Removing them would distort the analysis and potentially lead to inaccurate conclusions regarding which regions generate the highest sales or profit.

#### Explaining the Impact of Outliers on Analysis
**Effect on Statistical Measures**

Outliers can distort statistical metrics by inflating the mean and increasing the standard deviation. Once appropriately addressed, the mean and measures of spread become more stable and better represent the majority of the data.

**Effect on Machine Learning Models**

Outliers may bias predictive models, such as linear regression, toward extreme values. Handling outliers through capping or removal improves model generalizability and predictive accuracy.

**Effect on Visualizations**

Outliers can stretch visualizations, such as histograms and boxplots, compressing the majority of data points and obscuring patterns. Proper treatment of outliers allows for clearer and more interpretable visual representations.
